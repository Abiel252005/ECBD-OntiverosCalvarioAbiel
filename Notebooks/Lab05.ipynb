```python
df.columns = df.columns.str.strip().str.lower()
```

```python
print(df.isnull().sum())
```

```python
(df.isnull().sum() / len(df)) * 100
```

```python
df.duplicated().sum()
```

```python
# Eliminar columna director
df.drop(columns=["director"], inplace=True)
```

```python
df.columns
```

```python
df["country"] = df["country"].fillna(df["country"].mode()[0])
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df["date_added"] = df["date_added"].fillna(df["date_added"].mode()[0])
```

```python
df["duration"] = df["duration"].fillna(df["duration"].mode()[0])
```

```python
df["cast"] = df["cast"].fillna("Desconocido")
```

```python
# Verificamos que ya no queden valores nulos
df.isnull().sum()
```

```python
#Justificación
#Justificación: Se eliminó director por tener 30% de nulos para evitar problemas; country, rating, date_added y duration se rellenaron con la moda para conservar filas al faltar pocos datos, y cast se rellenó con 'Desconocido' por ser nombres propios donde la moda no aplica.
```

```python
#3. Analisis descriptivo (CAPA DE INFORMACIÓN)

#Graficamos el total de contenido por tipo
plt.figure(figsize=(10,5)) #tamaño
sns.countplot(df, x="type")
plt.title("Total de contenido por tipo")
plt.xlabel("Tipo (Movie / TV Show)")
plt.ylabel("Total")
plt.show()
```

```python
#Graficamos la clasificación (rating) por tipo de contenido
plt.figure(figsize=(12,5)) #tamaño
sns.countplot(df, x="rating", hue="type")
plt.title("Clasificación de contenido por tipo")
plt.xlabel("Clasificación (Rating)")
plt.ylabel("Total")
plt.xticks(rotation=45) # Rotamos las etiquetas para que no se encimen
plt.show()
```

```python
#Mostrar el año de lanzamiento por tipo de contenido
plt.figure(figsize=(8,5)) #tamaño
df_reciente = df[df['release_year'] > 2000]
sns.violinplot(df_reciente, x="type", y="release_year")
plt.title("Rango de año de lanzamiento por tipo (Post 2000)")
plt.xlabel("Tipo")
plt.ylabel("Año de Lanzamiento")
plt.show()
```

```python
#Graficamos los 10 países con más contenido
plt.figure(figsize=(12,5)) #tamaño
top_paises = df['country'].value_counts().head(10).index
sns.countplot(df[df['country'].isin(top_paises)], x="country", hue="type")
plt.title("Total de contenido por Top 10 Países y Tipo")
plt.xlabel("País")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.show()
```

1. La primera gráfica revela que Netflix tiene un catálogo dominado principalmente por Películas (Movies), superando por un margen amplio a las Series (TV Shows).


2. Según la distribución de los ratings, la plataforma concentra sus producciones en público maduro y adolescente (clasificaciones TV-MA y TV-14), dejando en segundo plano el contenido exclusivamente infantil.
   
4. El gráfico de violín muestra que la inmensa mayoría del contenido actual se produjo después del año 2015, indicando un acelerado crecimiento reciente. Además, Estados Unidos domina la creación de ambos formatos (películas y series) a nivel global.
