```python
import pandas as pd
```

```python
#Crear un dataFRame con 1,000,000 de registros
cantidad=1000000
```

```python
import random
```

```python
influencers = ["Fernanfloo", "Vegueta77", "Gael_Cordova", "Betito", "Amlo", "Donald Trump", "Quin_jon_hun", "Elonk Musk", "Vladimir Putin"]
tipos_post = ["Imagen", "Reel", "Video", "Carrusel"]
```

```python
df_instagram = pd.DataFrame({
    "influencer": [random.choice(influencers) for _ in range(cantidad)],
    "likes": [random.randint(1000, 575000) for _ in range(cantidad)],
    "comentarios": [random.randint(1000, 50000)for _ in range(cantidad)],
    "fecha": [random.randint(2020, 2026) for _ in range(cantidad)],
    "tipo_post": [random.choice(tipos_post) for _ in range(cantidad)]
})
```
