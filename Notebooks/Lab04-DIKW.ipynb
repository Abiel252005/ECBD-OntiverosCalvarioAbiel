```python
#Limpieza
#Nombres de columna

df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#Eliminar colummna cabin

df.drop(columns=["cabin"], inplace=True)
```

```python
df.columns
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos la columna Fare con el valor mas repetido (moda)

df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Validar datos
df["age"].min()
```

```python
df["age"].max()
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(10,5)) #tamaño
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.show()
```

```python
plt.figure(figsize=(8,4)) #tamaño
sns.countplot(df, x="sex", hue="pclass")
plt.title("Total de Genero por clase")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
#Mostrar la edad por genero y clase
plt.figure(figsize=(6,4))
sns.violinplot(df,x="sex", y="age", hue="pclass")
plt.title("Rango de edad por genero y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMINETO
#Los pasajeros de primera clase pagaban tarifas mas altas
df.groupby("pclass")["fare"].mean()
```
